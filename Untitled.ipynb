```python
import pandas as pd
df = pd.read_excel('dataset.xlsx')
df
```

```python
from sklearn.model_selection import train_test_split 

x = df[['No of Stations', 'No of Slots', 'Beacon Interval']]
y = df[['Throughput']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

df_norm = (df - df.mean())/(df.max() - df.min())
x_norm = df_norm[['No of Stations', 'No of Slots', 'Beacon Interval']]
y_norm = df_norm[['Throughput']]
x_norm_train, x_norm_test, y_norm_train, y_norm_test = train_test_split(x_norm, y_norm, test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeRegressor

regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1_norm = DecisionTreeRegressor(max_depth=2)
regr_2_norm = DecisionTreeRegressor(max_depth=5)

regr_1.fit(x_train, y_train)
regr_2.fit(x_train, y_train)
regr_1_norm.fit(x_norm_train, y_norm_train)
regr_2_norm.fit(x_norm_train, y_norm_train)
```

```python
y_1 = regr_1.predict(x_test)
y_2 = regr_2.predict(x_test)
y_1_norm = regr_1_norm.predict(x_norm_test)
y_2_norm = regr_2_norm.predict(x_norm_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_1)
```

```python
r2_score(y_test, y_2)
```

```python
r2_score(y_norm_test, y_1_norm)
```

```python
r2_score(y_norm_test, y_2_norm)
```

```python
regr_3_norm = DecisionTreeRegressor(max_depth=10)
regr_3_norm.fit(x_norm_train, y_norm_train)
y_3_norm = regr_3_norm.predict(x_norm_test)
```

```python
r2_score(y_norm_test, y_3_norm)
```

```python
regr_4_norm = DecisionTreeRegressor(max_depth=20)
regr_4_norm.fit(x_norm_train, y_norm_train)
y_4_norm = regr_4_norm.predict(x_norm_test)
```

```python
r2_score(y_norm_test, y_4_norm)
```
